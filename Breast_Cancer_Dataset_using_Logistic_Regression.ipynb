```python
import  pandas as pd
import numpy as np
import sklearn.datasets
from sklearn.preprocessing import StandardScaler # to standarise dataset
from sklearn.model_selection import train_test_split #to split data into train data nd test data
from sklearn.linear_model import LogisticRegression # we are going to use this modle so importing this model
from sklearn.metrics import accuracy_score # to display accararcy board
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
# standraization data
print(breast_cancer_dataset.data.mean())
print(breast_cancer_dataset.data.std())
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(breast_cancer_dataset.data)
```

```python
standerized_data = scaler.transform(breast_cancer_dataset.data)
```

```python
print(standerized_data)
```

```python
print(standerized_data.mean())
print(standerized_data.std())
```

```python
std_df = pd.DataFrame(standerized_data,columns=breast_cancer_dataset.feature_names)
```

```python
std_df.head()
```

```python
X = standerized_data
Y = breast_cancer_dataset.target
```

```python
print(X)
print(Y)
```

```python
# splinting data for test nd train
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape,X_test.shape,X_train.shape)
```

Model Training



```python
model = LogisticRegression()
```

```python
#training the logistic regression model with traing data
model.fit(X_train,Y_train)
```

Model Evaluation

find the accuary score

```python
# accuarcy on training data
X_train_prediction = model.predict(X_train)
training_data_accurarcy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accurarcy on training data : ",training_data_accurarcy )
```

```python
# accuarcy on training data
X_test_prediction = model.predict(X_test)
test_data_accurarcy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accurarcy on test data : ", test_data_accurarcy)
```
